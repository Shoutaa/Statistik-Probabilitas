```python
#dekomposisi R, G, B dari citra berwarna
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Step 1: Load an image (Replace the path with the actual image file path)
image_path = '/content/drive/MyDrive/SEMESTER 5/Pengolahan CItra Digital/Pertemuan 4/kota.jpg'  # Ganti dengan path gambar Anda
image = cv2.imread(image_path)

# Step 2: Convert the image from BGR (OpenCV default) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 3: Split the image into its Red, Green, and Blue channels
R, G, B = cv2.split(image_rgb)

# Step 4: Create a zero matrix to help in visualization
zeros = np.zeros(image.shape[:2], dtype="uint8")

# Step 5: Visualize the R, G, B channels separately
plt.figure(figsize=(10, 10))

# Red channel
plt.subplot(2, 2, 1)
red_image = cv2.merge([R, zeros, zeros])  # R channel with G, B as 0
plt.imshow(red_image)
plt.title('MERAH')
plt.axis('off')

# Green channel
plt.subplot(2, 2, 2)
green_image = cv2.merge([zeros, G, zeros])  # G channel with R, B as 0
plt.imshow(green_image)
plt.title('HIJAU')
plt.axis('off')

# Blue channel
plt.subplot(2, 2, 3)
blue_image = cv2.merge([zeros, zeros, B])  # B channel with R, G as 0
plt.imshow(blue_image)
plt.title('BIRU')
plt.axis('off')

# Original image for comparison
plt.subplot(2, 2, 4)
plt.imshow(image_rgb)
plt.title('GAMBAR ASLI')
plt.axis('off')

plt.show()
```

## Program 2

```python
#Uniform Mapping dan Logarithmic Mapping
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Step 1: Load an image (Replace the path with the actual image file path)
image_path = '/content/drive/MyDrive/SEMESTER 5/Pengolahan CItra Digital/Pertemuan 4/kota.jpg'  # Ganti dengan path gambar Anda
image = cv2.imread(image_path)

# Step 2: Convert the image from BGR (OpenCV default) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 3: Split the image into its Red, Green, and Blue channels
R, G, B = cv2.split(image_rgb)

# Step 4: Create a zero matrix to help in visualization
zeros = np.zeros(image.shape[:2], dtype="uint8")

# Uniform Mapping (Normalize intensity between 0 and 255)
def uniform_mapping(channel):
    return cv2.normalize(channel, None, 0, 255, cv2.NORM_MINMAX)

# Logarithmic Mapping
def logarithmic_mapping(channel):
    c = 255 / np.log(1 + np.max(channel))
    log_mapped = c * (np.log(channel + 1))
    return np.array(log_mapped, dtype=np.uint8)

# Step 5: Apply Uniform and Logarithmic Mapping to each channel
R_uniform = uniform_mapping(R)
G_uniform = uniform_mapping(G)
B_uniform = uniform_mapping(B)

R_log = logarithmic_mapping(R)
G_log = logarithmic_mapping(G)
B_log = logarithmic_mapping(B)

# Step 6: Visualize the R, G, B channels separately and with Uniform/Logarithmic mappings
plt.figure(figsize=(15, 15))

# Original Red channel
plt.subplot(3, 3, 1)
red_image = cv2.merge([R, zeros, zeros])  # R channel with G, B as 0
plt.imshow(red_image)
plt.title('MERAH (Asli)')
plt.axis('off')

# Uniform Red
plt.subplot(3, 3, 2)
uniform_red_image = cv2.merge([R_uniform, zeros, zeros])
plt.imshow(uniform_red_image)
plt.title('MERAH (Uniform Mapping)')
plt.axis('off')

# Logarithmic Red
plt.subplot(3, 3, 3)
log_red_image = cv2.merge([R_log, zeros, zeros])
plt.imshow(log_red_image)
plt.title('MERAH (Logarithmic Mapping)')
plt.axis('off')

# Original Green channel
plt.subplot(3, 3, 4)
green_image = cv2.merge([zeros, G, zeros])  # G channel with R, B as 0
plt.imshow(green_image)
plt.title('HIJAU (Asli)')
plt.axis('off')

# Uniform Green
plt.subplot(3, 3, 5)
uniform_green_image = cv2.merge([zeros, G_uniform, zeros])
plt.imshow(uniform_green_image)
plt.title('HIJAU (Uniform Mapping)')
plt.axis('off')

# Logarithmic Green
plt.subplot(3, 3, 6)
log_green_image = cv2.merge([zeros, G_log, zeros])
plt.imshow(log_green_image)
plt.title('HIJAU (Logarithmic Mapping)')
plt.axis('off')

# Original Blue channel
plt.subplot(3, 3, 7)
blue_image = cv2.merge([zeros, zeros, B])  # B channel with R, G as 0
plt.imshow(blue_image)
plt.title('BIRU (Asli)')
plt.axis('off')

# Uniform Blue
plt.subplot(3, 3, 8)
uniform_blue_image = cv2.merge([zeros, zeros, B_uniform])
plt.imshow(uniform_blue_image)
plt.title('BIRU (Uniform Mapping)')
plt.axis('off')

# Logarithmic Blue
plt.subplot(3, 3, 9)
log_blue_image = cv2.merge([zeros, zeros, B_log])
plt.imshow(log_blue_image)
plt.title('BIRU (Logarithmic Mapping)')
plt.axis('off')

plt.show()
```

#ANALISIS DAN KESIMPULAN

##Analisis

**A. ANALISIS PROGRAM PERTAMA**

Program pertama bertujuan untuk dekomposisi saluran warna (Red, Green, dan Blue) dari gambar berwarna menggunakan OpenCV dan menampilkannya secara terpisah. Berikut langkah-langkah yang diambil dalam program ini:
1. **Membaca Gambar**: Gambar di-load menggunakan `cv2.imread()` dan dikonversi dari format BGR (default OpenCV) ke RGB.
2. **Memisahkan Saluran Warna**: Gambar kemudian dipisah ke dalam tiga saluran warna: R (Red), G (Green), dan B (Blue) menggunakan `cv2.split()`.
3. **Visualisasi**: Untuk setiap saluran (R, G, B), gambar yang hanya menampilkan salah satu saluran ditampilkan, sementara dua saluran lainnya diatur menjadi nol.

Output dari program ini memberikan gambaran visual bagaimana masing-masing saluran warna (R, G, B) berkontribusi terhadap keseluruhan gambar. Ini berguna untuk memahami komposisi warna dari gambar secara lebih mendetail.

**B. ANALISIS PROGRAM KEDUA**

Program kedua memperluas program pertama dengan melakukan **Uniform Mapping** dan **Logarithmic Mapping** pada masing-masing saluran warna (R, G, B).
1. **Uniform Mapping**: Fungsi ini digunakan untuk meratakan intensitas piksel pada rentang 0 hingga 255. Mapping ini memastikan distribusi intensitas merata.
2. **Logarithmic Mapping**: Fungsi ini menerapkan pemetaan logaritmik untuk memperbaiki kontras pada area yang memiliki rentang intensitas kecil. Metode ini memperjelas detail di area yang gelap tanpa terlalu mengubah intensitas keseluruhan gambar.

Program ini juga menampilkan hasil pemetaan uniform dan logaritmik untuk setiap saluran warna, memungkinkan perbandingan antara gambar asli dan hasil mapping. **Uniform mapping** lebih cocok untuk distribusi kontras yang merata, sementara **logarithmic mapping** membantu dalam meningkatkan kontras pada area gambar yang gelap.

Kesimpulan:
- **Program pertama** fokus pada pemisahan dan visualisasi saluran warna (R, G, B) yang sederhana, namun efektif untuk analisis dasar gambar berwarna.
- **Program kedua** memperkenalkan teknik pemetaan intensitas untuk meningkatkan kualitas visual gambar, seperti uniform dan logarithmic mapping. Kedua teknik ini dapat memperbaiki distribusi kontras dan detail dalam gambar.
